```python
import pandas as pd
```

# Resource shortage detection

```python
service_delivery = pd.read_csv("../Data/service_delivery.csv")
```

```python
service_delivery.head(10)
```

```python
service_delivery["shortage_score"] = service_delivery[
    ['Identify Shortages of Medical Supplies(Syringes)?',
     'Identify Shortages of Medical Supplies(Bandages)?',
     'Identify Shortages of Medical Supplies(Personal Protective Equipment)?']
].apply(lambda x: sum(x == "Yes"), axis=1)
```

```python
service_delivery['alert_level'] = service_delivery['shortage_score'].apply(lambda x: 'Low' if x == 0 else 'Medium' if x <= 2 else 'High')
```
